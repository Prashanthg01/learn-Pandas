```python
import pandas as pd
```

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Value': [10, 20, 30, 40, 50, 60]}
df = pd.DataFrame(data)
display(df)
# Calculate the sum of 'Value' for each category
result_sum = df.groupby('Category')['Value'].sum()
print("Sum:")
print(result_sum)

# Calculate the mean of 'Value' for each category
result_mean = df.groupby('Category')['Value'].mean()
print("\nMean:")
print(result_mean)

# Calculate the minimum of 'Value' for each category
result_min = df.groupby('Category')['Value'].min()
print("\nMin:")
print(result_min)

# Calculate the maximum of 'Value' for each category
result_max = df.groupby('Category')['Value'].max()
print("\nMax:")
print(result_max)

# Calculate the count of values for each category
result_count = df.groupby('Category')['Value'].count()
print("\nCount:")
print(result_count)
```

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Value1': [10, 20, 30, 40, 50, 60],
        'Value2': [100, 200, 300, 400, 500, 600]}
df = pd.DataFrame(data)
display(df)
# Group by 'Category' and calculate the sum of 'Value1' and 'Value2' for each category
result_grouped_sum = df.groupby('Category').sum()
print("Grouped Sum:")
print(result_grouped_sum)
```
