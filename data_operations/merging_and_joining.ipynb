```python
import pandas as pd
```

```python
# Creating two DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})

display(df1)
display(df2)
# Merge based on 'key' column
merged_df = pd.merge(df1, df2, on='key', how='inner')
display(merged_df)

# try with different type of join: ('inner', 'outer', 'left', 'right').
```

```python
# Creating two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]}, index=['a', 'b', 'd'])

display(df1)
display(df2)
# Joining based on index
joined_df = df1.join(df2, how='inner')
display(joined_df)
```

```python
# Creating two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})

display(df1)
display(df2)
# Concatenating along rows
concatenated_df = pd.concat([df1, df2])
display(concatenated_df)
```

```python
import pandas as pd

# Creating two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})

display(df1)
display(df2)
# Merging
merged_df = pd.merge(df1, df2, on='A', how='inner')
display("Merge:")
display(merged_df)

# Joining
joined_df = df1.join(df2, how='inner', lsuffix='_df1', rsuffix='_df2')
display("\nJoin:")
display(joined_df)

# Concatenating
concatenated_df = pd.concat([df1, df2])
display("\nConcatenate:")
display(concatenated_df)
```
