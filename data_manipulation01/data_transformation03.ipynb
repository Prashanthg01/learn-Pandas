```python
import pandas as pd
```

```python
data = {'A': [1, 2, 3, 4],
        'B': [10, 20, 30, 40],
        'C': ["sol_gd", "lol_gog", "fof_mom", "sos_kok"]}
df = pd.DataFrame(data)
display(df)

# Apply a function to each column
df_squared = df.apply(lambda x: x**2 if x.dtype == 'int64' else x)
display(df_squared)

df['C'] = df['C'].apply(lambda x: x.split("_")[0])
display(df)
```

```python
data = {'Category': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)
display(df)

# Convert categorical variable into dummy/indicator variables
df_dummies = pd.get_dummies(df['Category'])
display(df_dummies)
```

```python
data = {'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02'],
        'Type': ['A', 'B', 'A', 'B'],
        'Value': [10, 20, 30, 40]}
df = pd.DataFrame(data)
display(df)

# Pivot the DataFrame
df_pivot = df.pivot(index='Date', columns='Type', values='Value')
display(df_pivot)
```

```python
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-02-04', '2022-02-08'],
        'Value': [10, 20, 30, 40, 10]}
df = pd.DataFrame(data)
display(df)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' column as index
df.set_index('Date', inplace=True)

# Resample the time series to monthly frequency
df_resampled = df.resample('M').sum()
print(df_resampled)
```
