```python
import pandas as pd
import numpy as np
```

```python
data = {'A': [1, np.nan, 3, 4],
        'B': ['a', 'b', np.nan, 'd']}
df = pd.DataFrame(data)
display(df)

# Drop rows with NaN values
df_dropped = df.dropna()
display(df_dropped)
```

```python
data = {'A': [1, np.nan, 3, 4],
        'B': ['a', 'b', np.nan, 'd']}
df = pd.DataFrame(data)
display(df)

# Fill NaN values with 0
df_filled = df.fillna(0)
display(df_filled)
```

```python
data = {'A': [1, np.nan, 3, 4],
        'B': ['a', 'b', np.nan, 'd']}
df = pd.DataFrame(data)
display(df)

# Interpolate missing values
df_interpolated = df.interpolate()
display(df_interpolated)
```
